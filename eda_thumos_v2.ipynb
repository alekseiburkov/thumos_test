#### Date range columns

```python
columns_to_dt = ['job_date_range',
                 'job_date_range2',
                 'school_date_range',
                 'school_date_range2']
```

```python
def parse_date(date_str):
    if 'present' in date_str:
        current_year = datetime.now().year
        return current_year
    else:
        years = re.findall(r'\d{4}', date_str)
        if years:
            return int(years[0])
    return None
```

```python
def calculate_difference(row):
    dates = row.split('â€“')
    if len(dates) == 2:
        start_year = parse_date(dates[0])
        end_year = parse_date(dates[1])
        if start_year is not None and end_year is not None:
            return end_year - start_year
    return None
```

```python
for column in columns_to_dt:
    df_eda[column] = df_eda[column].astype(str).apply(calculate_difference)
```

Let's make a function to categorize and standardize degrees.

```python
bachelor_list = ['bs',
                 'ba',
                 'b.s.',
                 'b.a.',
                 'bach',
                 'b.',
                 'b.s',
                 'b.a',
                 'bfa',
                 'b.f.a',
                 'bla']

masters_list = ['ms',
                'ma',
                'm.s.',
                'm.a',
                'master',
                'mfa',
                'mpp']

doctorate_list = ['doct',
                  'phd',
                  'ph.d',
                  'j.d']

mba_list = ['mba']

ongoing_list = ['ong',
                'process']

college_list = ['assoc',
                'a.s',
                'ccp',
                'aa']
```

```python
def degree_stand(row):
    if any(keyword in row for keyword in mba_list):
        return 'mba'
    elif any(keyword in row for keyword in doctorate_list):
        return 'doctor'
    elif any(keyword in row for keyword in ongoing_list):
        return 'ongoing'
    elif any(keyword in row for keyword in college_list):
        return 'college'
    elif any(keyword in row for keyword in bachelor_list):
        return 'bachelor'
    elif any(keyword in row for keyword in masters_list):
        return 'master'
    else: return 'other'
```

```python
df_eda['school_degree'] = df_eda['school_degree'].apply(degree_stand)
df_eda['school_degree2'] = df_eda['school_degree2'].apply(degree_stand)
```

#### Result

We need to understand what features make a person more likely to purchase a membership. Let's make result a boolean feature that tells us if the person did purchase or is already a member or is not.

Note: There can be a further study to determine features for different outcomes (approved seems better that out of sequence), but for the purposes of this test I decided not to subdivide it due to time constarints.

```python
df_eda['result'].value_counts()
```

```python
keywords_member = ['purchased',
                   'member']

keywords_not = ['sent personal',
               'approved',
               'out of sequence']
```

```python
def cat_result(row):
    if any(keyword in row for keyword in keywords_member):
        return 1
    elif any(keyword in row for keyword in keywords_not):
        return 0
    else: return 'Error'
```

```python
df_eda['result'] = df_eda['result'].apply(cat_result)
```

```python
df_eda['result'].value_counts()
```

## Model

Since we have a lot of categorical values in our features to understand importance of the fetures let's build a model to predict the outcome and look at feature importance.

```python
df_model = df_eda.drop('linkedin_profile', axis=1)
```

```python
df_model.info()
```

```python
features = df_model.drop('result', axis=1)
target = df_model['result']

display(features.head())
print(target.head())
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target,
    test_size=0.25,
    random_state=RANDOM_STATE,
    stratify=target)
```

```python
print(features.shape)
print(features_train.shape, features_test.shape)
print(target_train.shape, target_test.shape)
```

```python
features.columns
```

```python
columns_to_encode = ['email',
                     'description',
                     'headline',
                     'location',
                     'mail_from_dropcontact',
                     'company',
                     'job_title',
                     'job_description',
                     'job_location',
                     'company2',
                     'job_title2',
                     'job_description2',
                     'school',
                     'school_degree',
                     'school_degree_spec',
                     'school2',
                     'school_degree2',
                     'school_degree_spec2',
                     'qualification_from_drop_contact',
                     'civility_from_drop_contact',
                     'phone_number_from_drop_contact',
                     'website_from_drop_contact',
                     'twitter',
                     'website',
                     'company_website',
                     'skill1',
                     'skill2',
                     'skill3',
                     'skill4',
                     'skill5',
                     'skill6',
                     'job_location2',
                     'school_description2',
                     'naf5_code_from_drop_contact',
                     'naf5_des_from_drop_contact',
                     'siren_from_drop_contact',
                     'siret_from_drop_contact',
                     'siret_address_from_drop_contact',
                     'siret_zip_from_drop_contact',
                     'vat_from_drop_contact',
                     'school_description',
                     'nb_employees_from_drop_contact']
```

```python
model = CatBoostClassifier(random_state=RANDOM_STATE,
                           cat_features=columns_to_encode,
                           verbose=500,
                           auto_class_weights='Balanced')
```

```python
model.fit(features_train, target_train)

prediction_cbc = model.predict(features_test)
probabilities_cbc = model.predict_proba(features_test)
probabilities_one_cbc = probabilities_cbc[:, 1]
```

```python
roc_auc = roc_auc_score(target_test, probabilities_one_cbc)
print('ROC-AUC', roc_auc)
```

Very good ROC-AUC score.

```python
feature_importance = model.get_feature_importance(type=catboost.EFstrType.FeatureImportance)
```

```python
importance_df = pd.DataFrame({'Feature': features.columns, 'Importance': feature_importance}).sort_values(by='Importance', ascending=False).reset_index(drop=True)
```

Top 15 most important features for predictiong result.
